### TEXT VETORIZATION
This is the last step to prepare data for machine learning, This is a numerical reprentation that machine learning can understand 

There are two main techniques in this course
- Bag of words: Counts which words appear in which documents 
- Term Frequency Inverse Document Frequency (TF DIF): calculates the importance of that word for that particular document, and takes into account how that word also appears in each of the other documents

````python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer # a tool for text processing, it transforms a collection of text documents into a matrix of token counts
````

````python
data = [' Most shark attacks occur about 10 feet from the beach since that is where the people are',
        'the efficiency with which he paired the socks in the drawer was quite admirable',
        'carol drank the blood as if she were a vampire',
        'giving directions that the mountains are to the west only works when you can see them',
        'the sign said there was road work ahead so he decided to speed up',
        'the gruff old man sat in the back of the bait shop grumbling to himself as he scooped out a handful of worms']
````

````python
countvec = CountVectorizer()
````

````python
countvec_fit = countvec.fit_transform(data) 
# fit_transofrm looks at the text and find unique words
# Converts the text into numbers and counts how often each word occurs
````

````python
# Every row = a document
# Every Column = a word from the vocabulary
# Every Entry = a word count
# countvec.get_feature_names_out() returns the actual words in the vocabulary --> column headers
bag_of_words = pd.DataFrame(countvec_fit.toarray(), columns=countvec.get_feature_names_out())
````

````python
print(bag_of_words)
````

### TF-IDF = Term Frequency × Inverse Document Frequency
#### Term Frequency (TF)
TF tells us how often a word appears in a document.
TF(word,document)=count of word in document / total words in document

```
In document "I love NLP NLP"

count of NLP = 2
total words = 4

So:
TF(NLP)=4/2=0.5
```
#### Inverse Document Frequency (IDF)
DF tells us how rare a word is across the whole corpus.
IDF(word) = log (N / DF(word))
- N = total number of documents
- DF(word) = number of documents containing that word
#### TF-IDF formula
TF-IDF(word,document)=TF(word,document)×IDF(word)
So:
- high TF = word appears a lot in this document
- high IDF = word is rare in the whole corpus
- together = word is important for this specific document

````python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
data = [' Most shark attacks occur about 10 feet from the beach since that is where the people are',
        'the efficiency with which he paired the socks in the drawer was quite admirable',
        'carol drank the blood as if she were a vampire',
        'giving directions that the mountains are to the west only works when you can see them',
        'the sign said there was road work ahead so he decided to speed up',
        'the gruff old man sat in the back of the bait shop grumbling to himself as he scooped out a handful of worms']
````

````python
tfidfvec = TfidfVectorizer()
````

````python
tfidfvec_fit = tfidfvec.fit_transform(data)
````

````python
tfidf_bag = pd.DataFrame(tfidfvec_fit.toarray(), columns = tfidfvec.get_feature_names_out())
````

````python
print(tfidf_bag)
````
