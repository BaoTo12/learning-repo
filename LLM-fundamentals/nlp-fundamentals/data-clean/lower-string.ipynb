List Comprehension
Syntax Pattern: [ expression for item in iterable if condition] 

```python
sentence_list = ["Could you pass me the RV remote", "It is IMPOSSIBLE to find this hotel", "Want to go for dinner on Tuesday?"]
```

```python
lower_sentence_list = [ str.lower() for str in sentence_list ]
```

```python
print(lower_sentence_list)
```
