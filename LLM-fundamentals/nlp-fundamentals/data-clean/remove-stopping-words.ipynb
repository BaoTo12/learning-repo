```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
sentence = "it is too far to go to the shop and he did not want to go there"
```

```python
en_stopwords.append("go")
```

```python
sentence_no_stopwords = " ".join([x for x in sentence.split() if x not in en_stopwords])
print(sentence_no_stopwords)
```
