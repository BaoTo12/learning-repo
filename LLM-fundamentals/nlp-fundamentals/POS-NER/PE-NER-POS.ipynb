```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize  import word_tokenize
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot  as plt
```

### Load data

```python
bbc_data = pd.read_csv("bbc-news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data["title"])
titles.head()
```

### Clean data

```python
# Lower Case
titles["lower_case"] = titles["title"].str.lower()
titles.head()
```

```python
# stop word removal
en_stopwords = stopwords.words("english")
titles['no_stopwords'] = titles["lower_case"].apply(lambda x: " ".join([word for word in x.split() if word not in en_stopwords]))
titles.head()
```

```python
# punctation removal
titles['no_stopwords_no_punct'] = titles['no_stopwords'].apply(lambda x: re.sub(r"([^\w\s])", "", x))
titles.head()
```

```python
# tokenize
titles["tokens_raw"] = titles.apply(lambda x: word_tokenize(x["title"]), axis = 1)
titles["tokens_clean"] = titles["no_stopwords_no_punct"].apply(lambda x: word_tokenize(x))
titles.head()
```

```python
# lemmatizing
lemmatizer = WordNetLemmatizer()
titles["tokens_clean_lemmatized"] = titles["tokens_clean"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
# create lists for just our tokens
tokens_raw_list = sum(titles["tokens_raw"], [])
tokens_clean_list = sum(titles['tokens_clean_lemmatized'], [])
```

```python
tokens_clean_list
```

### POS Tagging

```python
nlp = spacy.load("en_core_web_sm") # pre trained model for text processing
```

```python
spacy_doc = nlp(" ".join(tokens_raw_list)) # run text through  multiple NLP tasks: tokenizer, POS tagging, NER
```

```python
# extract the tokens and pos tags into a dataframe
pos_df = pd.DataFrame(columns=["token", "pos_tag"])

for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{"token": token.text, "pos_tag": token.pos_}])], ignore_index=True)
```

```python
pos_df
```

```python
# token frequency count
pos_df_counts = pos_df.groupby(["token", "pos_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
pos_df_counts.head(10)
```

```python
# most common nouns
nouns = pos_df_counts[pos_df_counts["pos_tag"] == "NOUN"][0:10]
nouns
```

```python
# most common verbs
verbs = pos_df_counts[pos_df_counts.pos_tag == "VERB"][0:10]
verbs
```

```python
# most common adjectives
adj = pos_df_counts[pos_df_counts.pos_tag == "ADJ"][0:10]
adj
```

### NER

```python
# extract the tokens and entity tags into a dataframe
ner_df = pd.DataFrame(columns=["token", "ner_tag"])
for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
      ner_df = pd.concat([ner_df, pd.DataFrame.from_records([{"token": token.text, "ner_tag": token.label_}])], ignore_index=True)  
    
```

```python
ner_df.head()
```

```python
# token frequency count
ner_df_counts = ner_df.groupby(['token','ner_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
ner_df_counts.head(10)
```

```python
# most common people
people = ner_df_counts[ner_df_counts.ner_tag == "PERSON"][0:10]
people
```

```python
# most common places
places = ner_df_counts[ner_df_counts.ner_tag == "GPE"][0:10]
places
```
