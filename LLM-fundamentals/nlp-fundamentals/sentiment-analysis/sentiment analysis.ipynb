### Rule-Based Sentiment Analysis 
is a method that uses a set of predefined rules to connect words to emotions or attitudes
Each word is given a **Polarity Score**, which is a numerical value that shows whether the word is positive, negative or neutral -1...0...1

#### DrawBacks
- Cannot handle subtle expressions, such as sarcasm, irony

#### Python packages that can handle rule-based sentiment analysis
- TextBlob (from textblob import TextBlob)
- Vader (from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer)

each package carries its own lexicon, which is a collection of words paired with their predefined sentiment scores

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

### TextBlob

```python
from textblob import TextBlob
```

```python
print(sentence_1)
sentiment_score_1 = TextBlob(sentence_1)
print(sentiment_score_1.sentiment.polarity)
```

```python
print(sentence_2)
sentiment_score_2 = TextBlob(sentence_2)
print(sentiment_score_2.sentiment.polarity)
```

```python
print(sentence_3)
sentiment_score_3 = TextBlob(sentence_3)
print(sentiment_score_3.sentiment.polarity)
```

```python
print(sentence_4)
sentiment_score_4 = TextBlob(sentence_4)
print(sentiment_score_4.sentiment.polarity)
```

### VADER

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
print(sentence_1)
print(vader_sentiment.polarity_scores(sentence_1))
```

```python
print(sentence_2)
print(vader_sentiment.polarity_scores(sentence_2)) 
```

### Pre Trained Transformers

Transformers can capture how words influence and relate to each other even when they appear far aprat in a sentence
It can understand context

```python
import transformers
from transformers import pipeline
```

```python
sentiment_analysis = pipeline("sentiment-analysis")
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
print(sentence_1)
sentiment_analysis(sentence_1)
```

```python
# We can choose another model that is trained by a different datasets
specific_model = pipeline("sentiment-analysis", model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
print(sentence_1)
specific_model(sentence_1)
```
