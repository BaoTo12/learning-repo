Tokenization means breaking text into smaller units (token)
Depends on your use cases, token can be word, sentence, ...

```python
import nltk 
nltk.download("punkt_tab") 
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = "Her cat's name is Luna. Her dog's name is max"
```

```python
sent_tokenize(sentences)
```

```python
sentence = "Her cat's name is Luna"
```

```python
word_tokenize(sentence)
```

```python
sentence_2 = "Her cat's name is Luna and her dog's name is max"
```

```python
result = word_tokenize(sentence_2)
print(result)
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
en_stopwords = stopwords.words("english")
```

```python
result_nostopwords = [ x for x in result if x not in en_stopwords ]
```

```python
print(result_nostopwords)
```
