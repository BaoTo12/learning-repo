Stemmingm means reducing words into their base form

```python
from nltk.stem import PorterStemmer 
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
```

```python
for t in connect_tokens:
    print(t, ": ", ps.stem(t))
```
